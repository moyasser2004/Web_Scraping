```python
def get_phone_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    
    phone_name = soup.find("div", attrs={"class": "-pls -prl"}).text
    phone_price = soup.find("div", attrs={"class": "df -i-ctr -fw-w"}).text.split(" ")[1]
    phone_branch = soup.select_one(".-phs .-pvxs a").text.strip()
    phone_in_stock = soup.find("p", attrs={"class": "-df -i-ctr -fs12 -pbs -rd5"}).text
    phone_delivery_state = soup.find("div", attrs={"class": "markup -fs12 -pbs"}).text
    phone_rating = soup.find("div", attrs={"class": "stars _m _al"}).text.split(" ")[0]
    phone_promotions = soup.find("div", attrs={"class": "-pvxs -hr _bet"}).text
    phone_images = list(
    map(
        lambda x: x["href"],
        soup.find("div", attrs={"class": "sldr _img _prod -rad4 -oh -mbs"}).findAll(
            "a"
        ),
    ),
)
    phone_details = (
        soup.find("div", attrs={"class": "markup -mhm -pvl -oxa -sc"})
        .find("table")
        .find("tbody")
        .findAll("tr")
    )
    phone_details = list(map(lambda x: x.text, phone_details))
    phone_details = "".join(phone_details)
    
    phone_customer_feedBack = soup.find(
        "div", attrs={"class": "-df -d-co -i-ctr -j-ctr -pvl"}
    ).text
    phone_color = soup.find("ul", attrs={"class": "-pvs -mvxs -phm -lsn"}).text.split(":")[2].strip()
    phone_feature = soup.find("div", attrs={"class": "markup -pam"}).text
    return{
        "phone_name": phone_name,
        "phone_price": phone_price,
        "phone_branch": phone_branch,
        "phone_in_stock": phone_in_stock,
        "phone_delivery_state": phone_delivery_state,
        "phone_rating": phone_rating,
        "phone_promotions": phone_promotions,
        "phone_images": phone_images,
        "phone_details": phone_details,
        "phone_customer_feedBack": phone_customer_feedBack,
        "phone_color": phone_color,
        "phone_feature": phone_feature
        
    }
```

```python
url = "https://www.jumia.com.eg/realme-realme-c51-128gb-4gb-mint-green-104701190.html"
get_phone_data(url)
```
